```python
X = vectorizer.fit_transform(final_df['SA_NOPUNCT']) #fit extracts the unique words from all the messages,transforms all messages to count vectors
X

```

```python
TF_vectors = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out()) #creating a dataframe 'pd'by representing transformed messages as array and column names=feature names

TF_vectors.head()
```

```python
TF_vectors.columns
```

```python
#scale the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler(with_mean= False) # create an instance of standard scalar

# transform data
X_scaled = scaler.fit_transform(TF_vectors)

X_scaled
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
#FINDING THE BEST K VALUE USING ELBOW METHOD
Sum_of_squared_distances = []
K = range(1,20)
for k in K:
   km = KMeans(n_clusters= k, init='k-means++',random_state=0)
   km = km.fit(X_scaled)
   Sum_of_squared_distances.append(km.inertia_) # . inertia will give the distance between the centriod to all other points in a cluster
   



plt.title('Elbow Method For Optimal k')
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum_of_squared_distances')
plt.show()
```

```python
# initialize kmeans with 2 centroids
kmeans = KMeans(n_clusters= 2, random_state= 42, init='k-means++')
```

```python
# fit the model

kmeans.fit(X_scaled)
```

```python
# store cluster labels in a variable
clusters = kmeans.labels_
clusters
```

```python
#Getting the Centroids
centroids = kmeans.cluster_centers_
centroids
```

##### Visualize the clusters using tSNE - t-distributed Stochastic Neighbor Embedding 

```python
from sklearn.manifold import TSNE
```

```python
#CREATING AN INSTANCE OF TSNE

tsne = TSNE(n_components=2, verbose=1, perplexity = 10, n_iter=3000,init='pca',random_state=42)

tsne_scores = tsne.fit_transform(X_scaled) 

tsne_scores
```

```python
# save our two dimensions into x0 and x1
Tx0 = tsne_scores[:, 0]
Tx1 = tsne_scores[:, 1]
```

```python
#Visualization of CLusters using tSNE and KMeans
import numpy as np

vis_df = pd.DataFrame() # Create a new dataframe
vis_df['X0'] = Tx0 # store first PCA component in X0
vis_df['X1'] = Tx1 # store second PCA component in X1
vis_df['clusters'] = clusters # Store KMeans clusters in clusters
vis_df['scores'] = final_df['Scores']
vis_df['C_pred_scores'] = final_df['C_Random_Forest_Predicted_Values']
vis_df['R_pred_scores'] = final_df['Reg_Random_Forest_Predicted_Values']


# Create a column of student number in a dataframe

vis_df['S'] = np.arange(len(vis_df))

vis_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.spatial import ConvexHull

plt.figure(figsize=(12, 7))

# Set axes names
plt.xlabel("TSNE C1", fontsize=16)
plt.ylabel("TSNE C2", fontsize=16)

# Create scatter plot with seaborn, where hue is the cluster assignment
ax = sns.scatterplot(data=vis_df, x='X0', y='X1', hue='clusters', palette=['green','orange'], s=300)
n_clusters = 2
# Plot cluster boundaries using convex hulls
for cluster_label in range(n_clusters):
    cluster_points = vis_df[vis_df['clusters'] == cluster_label][['X0', 'X1']]
    hull = ConvexHull(cluster_points)
    for simplex in hull.simplices:
        plt.plot(cluster_points.iloc[simplex, 0], cluster_points.iloc[simplex, 1], 'k--')

# Annotate each data point
for i, txt in enumerate(vis_df.S):
    ax.annotate(txt, (vis_df.X0.iat[i] + 0.09, vis_df.X1.iat[i]), fontsize=16)

# Increase the size of the legend
legend = ax.legend(title="Cluster", fontsize=14, title_fontsize=16)

# Show the plot
plt.show()
```

```python
# Initialize an empty list to store the HTML tables
html_tables = []

# Include the headings for each column only once
html_tables.append("""
<!DOCTYPE html>
<html>
<head>
</head>
<body>
<table border="1">
    <tr>
        <th>STDID</th>
        <th>Reference Answer</th>
        <th>Reference Answer Keywords</th>
        <th>Student Answer</th>
        <th>Student Answer Keywords</th>
        <th>Cosine Similarity</th>
        <th>Statistical Similarity</th>
        <th>Semantic Similarity</th>
        <th>Summary Similarity</th>
        <th>Actual Marks</th>
        <th>Predicted Marks</th>
    </tr>
                 
""")

# Iterate through the rows in selected_rows
for index, row in selected_rows.iterrows():
    stdid = row['STDID']  # Get the student ID
    refans = row['REFANS']
    stuans = row['STUANS']
    cosine_similarity = row['TFIDF_sim'] 
    Evaluator_Marks = row['Scores'] 
    Predicted_Marks = row['C_Random_Forest_Predicted_Values']
    # Calculate statistical similarity, semantic similarity, and summary similarity (replace with actual calculation)
    statistical_similarity = row['sta_sim']
    semantic_similarity = round(row['SEM_sim'], 2)  # Round to two decimal places
    summary_similarity = round(row['SUM_sim'], 2)  # Round to two decimal places



    # Tokenize, lowercase, remove stopwords, and lemmatize the reference answer and student answer
    ref_keywords = set(lemmatizer.lemmatize(word.lower()) for word in refans.split() if word.lower() not in stop_words)
    stu_keywords = set(lemmatizer.lemmatize(word.lower()) for word in stuans.split() if word.lower() not in stop_words)

    # Create a set of matched keywords (case-insensitive)
    matched_keywords = ref_keywords.intersection(stu_keywords)


    # Create HTML formatted strings with matched keywords in bold and green, and unmatched keywords in both answers in red
    formatted_refans_keywords = []
    formatted_stuans_keywords = []

    for keyword in ref_keywords:
        if keyword in matched_keywords:  # Matching keyword
            formatted_refans_keywords.append(f'<span style="color: green; font-weight: bold;">{keyword}</span>')
        else:  # Unmatched keyword in both answers
            formatted_refans_keywords.append(f'<span style="color: red; font-weight="bold">{keyword}</span>')

    for keyword in stu_keywords:
        if keyword in matched_keywords:  # Matching keyword
            formatted_stuans_keywords.append(f'<span style="color: green; font-weight: bold;">{keyword}</span>')
        else:  # Unmatched keyword in both answers
            formatted_stuans_keywords.append(f'<span style="color: red; font-weight: bold;">{keyword}</span>')

    # Join the keywords back into a single string
    highlighted_refans_keywords = ', '.join(formatted_refans_keywords)
    highlighted_stuans_keywords = ', '.join(formatted_stuans_keywords)

    # Create an HTML table for the pair of answers
    html_table = f"""
    <tr>
        <td>{stdid}</td>
        <td>{refans}</td>
        <td>{highlighted_refans_keywords}</td>
        <td>{stuans}</td>
        <td>{highlighted_stuans_keywords}</td>
        <td>{cosine_similarity}</td>
        <td>{statistical_similarity}</td>
        <td>{semantic_similarity}</td>
        <td>{summary_similarity}</td>
        <td>{Evaluator_Marks}</td>
        <td>{Predicted_Marks}</td>
    </tr>
    """

    # Append the HTML table to the list
    html_tables.append(html_table)

# Close the HTML content
html_tables.append("</table></body></html>")

# Join the HTML tables into a single HTML document
html_content = '\n'.join(html_tables)

# Save the HTML content to an HTML file
with open("Q1_IDEAS.html", "w", encoding="utf-8") as file:
    file.write(html_content)

print("HTML content saved to output.html")
```
